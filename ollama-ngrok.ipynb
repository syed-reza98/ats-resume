```python
# Cell 2: Run the Server and Ngrok
import subprocess
import time
import os
from pyngrok import ngrok
# from kaggle_secrets import UserSecretsClient
# user_secrets = UserSecretsClient()
NGROK_TOKEN = "3DMdHOb0lfnsNRDq2I3uA1EhK5O_6q8yphm79KgfBEzdss1iW"

# 1. Kill any existing Ollama process so the new one can bind port 11434
#    with OLLAMA_ORIGINS=* in effect. Without this step, re-running the
#    cell leaves the old (no-ORIGINS) process running and the new one fails
#    silently because the port is already in use.
print("Stopping any existing Ollama process...")
subprocess.run("pkill -f 'ollama serve' || true", shell=True)
time.sleep(2)

# 2. Start Ollama with OLLAMA_ORIGINS=* so requests forwarded through
#    ngrok are not blocked by Ollama's built-in CORS guard.
print("Starting Ollama server with OLLAMA_ORIGINS=*...")
subprocess.Popen(
    ["/usr/local/bin/ollama", "serve"],
    env={**os.environ, "OLLAMA_ORIGINS": "*"},
)
time.sleep(5)  # Give it a few seconds to boot up

# 2. Pull the model (also using absolute path)
print("Pulling Llama 3.1... this will take a minute or two.")
!/usr/local/bin/ollama pull llama3.1

# 3. Set up the Ngrok Tunnel
ngrok.set_auth_token(NGROK_TOKEN) # <-- REPLACE THIS

# Ollama runs on port 11434 by default
public_url = ngrok.connect(11434).public_url

print("\n" + "="*50)
print(f"✅ YOUR API ENDPOINT IS READY!")
print(f"👉 BASE URL: {public_url}/v1")
print("="*50 + "\n")

# Keep the notebook running
while True:
    time.sleep(60)
```
